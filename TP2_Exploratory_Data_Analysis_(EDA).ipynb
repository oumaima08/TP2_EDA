```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, Normalizer
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Read the CSV file into a DataFrame
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv("automobile.csv", names=headers)
```

```python
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head()
df.size
```

```python
avg_wheel_base = df["wheel-base"].astype("float").mean(axis = 0)
print("average wheel-base:", avg_wheel_base)
```

```python
avg_peak_rpm = df["peak-rpm"].astype("float").mean(axis = 0)
print("average peak-rpm:", avg_peak_rpm)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
df.replace({"horsepower": {np.nan: avg_horsepower}}, inplace=True)
```

```python
df.replace({"wheel-base": {np.nan: avg_wheel_base}}, inplace=True)
```

```python
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("average stroke", avg_stroke)
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)

#Cela indique que l'opération doit être effectuée sur les lignes (axis=0).
#Si vous vouliez supprimer des colonnes contenant des valeurs manquantes, vous utiliseriez axis=1.
```

```python
# transform mpg to L/100km by mathematical operation (235 divided by mpg)
df["highway-mpg"] = 235/df["highway-mpg"]

# rename column name from "highway-mpg" to "highway-L/100km"
df.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)

# check your transformed data
df.head()
```

```python
# check your transformed data
df.head(200)
```

```python
# get indicator variables of aspiration and assign it to data frame "dummy_variable_2"
dummy_variable_2 = pd.get_dummies(df['aspiration'])

# change column names for clarity
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=False)

# show first 5 instances of data frame "dummy_variable_2"
dummy_variable_2.head()
```

```python
# merge the new dataframe
df = pd.concat([df, dummy_variable_2], axis=1)

# drop original column "aspiration" from "df"
df.drop('aspiration', axis = 1, inplace=True)

df.to_csv('concat_df.csv')
df.head(200)
```
